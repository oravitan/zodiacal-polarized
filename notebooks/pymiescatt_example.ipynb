```python
import numpy as np
import matplotlib.pyplot as plt

from PyMieScatt import ScatteringFunction
```

```python
radius = 10000  # droplet size in nm
wavelength = 650  # wavelength in nm
water_ref_index = 1.332
```

```python
theta, SL, SR, SU = ScatteringFunction(water_ref_index, wavelength, 2*radius)
```

```python
theta_deg = np.rad2deg(theta)
plt.figure()
plt.semilogy(theta_deg, SL, label='SL')
plt.semilogy(theta_deg, SR, label='SR')
plt.semilogy(theta_deg, SU, label='SU')
plt.xlabel('theta (deg)')
plt.ylabel('Scattering amplitude')
plt.title('Water droplet scattering - all angles')
plt.grid()
plt.legend()
plt.show()
```

```python
# backscattering only
theta, SL, SR, SU = ScatteringFunction(water_ref_index, wavelength, 2*radius, minAngle=160, maxAngle=180, angularResolution=0.1)
```

```python
theta_deg = np.rad2deg(theta)
plt.figure()
plt.semilogy(theta_deg, SL, label='SL')
plt.semilogy(theta_deg, SR, label='SR')
plt.semilogy(theta_deg, SU, label='SU')
plt.xlabel('theta (deg)')
plt.ylabel('Scattering amplitude')
plt.title('Water droplet scattering - backscattering only')
plt.grid()
plt.legend()
plt.show()
```
